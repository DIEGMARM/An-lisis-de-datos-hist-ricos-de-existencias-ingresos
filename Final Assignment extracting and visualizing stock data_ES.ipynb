## Definir Función de Graficación


En esta sección, definimos la función `make_graph`. **No necesitas saber cómo funciona la función, solo debes preocuparte por las entradas. Toma un DataFrame con datos de acciones (el DataFrame debe contener las columnas Date y Close), un DataFrame con datos de ingresos (el DataFrame debe contener las columnas Date y Revenue), y el nombre de la acción.**


````python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Precio Histórico de la Acción", "Ingresos Históricos"), vertical_spacing = .3)
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name="Precio de la Acción"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date, infer_datetime_format=True), y=revenue_data_specific.Revenue.astype("float"), name="Ingresos"), row=2, col=1)
    fig.update_xaxes(title_text="Fecha", row=1, col=1)
    fig.update_xaxes(title_text="Fecha", row=2, col=1)
    fig.update_yaxes(title_text="Precio ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Ingresos ($US Millones)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
    from IPython.display import display, HTML
    fig_html = fig.to_html()
    display(HTML(fig_html))
````

## Pregunta 2: Usar Webscraping para Extraer Datos de Ingresos de Tesla


Usa la biblioteca `requests` para descargar la página web https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm Guarda el texto de la respuesta como una variable llamada `html_data`.


````python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
````

Analiza los datos HTML usando `beautiful_soup` con un parser, es decir, `html5lib` o `html.parser`.


````python
soup = BeautifulSoup(html_data, "html.parser")
````

Usando `BeautifulSoup` o la función `read_html`, extrae la tabla con los `Ingresos de Tesla` y guárdala en un DataFrame llamado `tesla_revenue`. El DataFrame debe tener las columnas `Date` y `Revenue`.


<details><summary>Instrucciones paso a paso</summary>

```

Aquí están las instrucciones paso a paso:

1. Crear un DataFrame Vacío
2. Encontrar la Tabla Relevante
3. Verificar la Tabla de Ingresos Trimestrales de Tesla
4. Iterar a Través de las Filas en el Cuerpo de la Tabla
5. Extraer Datos de las Columnas
6. Añadir Datos al DataFrame

```
</details>


<details><summary>Haz clic aquí si necesitas ayuda para localizar la tabla</summary>

```
    
A continuación se muestra el código para aislar la tabla, ahora necesitarás recorrer las filas y columnas como en el laboratorio anterior
    
soup.find_all("tbody")[1]
    
Si quieres usar la función read_html, la tabla está ubicada en el índice 1

Nos estamos enfocando en los ingresos trimestrales en el laboratorio.
```

</details>


````python
# Paso 1: Crear un DataFrame vacío
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Paso 2 Encontrar la tabla relevante
tables = soup.find_all("table")
print(f"Número de tablas encontradas: {len(tables)}")

# Inspeccionar cada tabla
for i, table in enumerate(tables):
    print(f"Tabla {i}:")
    print(table.prettify())

# 3: verificar que sea la de ingresos trimestrales de Tesla
# Según la pista, la tabla está en soup.find_all("tbody")[1]
table = soup.find_all("tbody")[1]

# Paso 4: Iterar a través de las filas en el cuerpo de la tabla
for row in table.find_all("tr"):  # Recorre todas las filas <tr>
    cols = row.find_all("td")  # Extrae todas las celdas <td> de la fila
    
    # Paso 5: Extraer datos de las columnas
    # Asumimos que la primera columna es Date y la segunda es Revenue
    if len(cols) >= 2:  # Verifica que haya al menos 2 columnas
        date = cols[0].text.strip()  # Fecha en la primera celda
        revenue = cols[1].text.strip()  # Ingresos en la segunda celda
        
        # Paso 6: Añadir datos al DataFrame
        tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})], ignore_index=True)

# Verificar el resultado
tesla_revenue.head()
````

Ejecuta la siguiente línea para eliminar la coma y el signo de dólar de la columna `Revenue`. 


````python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"",regex=True)
````

Ejecuta las siguientes líneas para eliminar valores nulos o cadenas vacías en la columna `Revenue`.


````python
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
````

````python
tesla_revenue.tail()
````
